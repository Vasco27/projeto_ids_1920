```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix

user_u = list(sorted(data.userId.unique()))
movie_u = list(sorted(data.movieId.unique()))

row = data.userId.astype('category', categories=user_u).cat.codes
col = data.movieId.astype('category', categories=movie_u).cat.codes

rate = data['rating'].tolist()

sparse_matrix = csr_matrix((rate, (row, col)), shape=(len(user_u), len(movie_u)))

df = pd.SparseDataFrame([ pd.SparseSeries(sparse_matrix[i].toarray().ravel(), fill_value=0) 
                              for i in np.arange(sparse_matrix.shape[0]) ], 
                       index=user_u, columns=movie_u, default_fill_value=0)
```

```python
pivot_table = data.pivot_table(index = "userId",columns = "title",values = "rating")
pivot_table.head(10)
```

```python
Average_ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
Average_ratings['Total Ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
Average_ratings.head(10)
```

```python
movie_watched = pivot_table["Invictus (2009)"]

correlations_zoolander = pivot_table.corrwith(movie_watched).sort_values(ascending=False)
correlations_zoolander.head()
```

```python
recommendation = pd.DataFrame(correlations_zoolander,columns=['Correlation'])
recommendation.dropna(inplace=True)
recommendation = recommendation.join(Average_ratings['Total Ratings'])
recommendation.head()
```

```python
recc = recommendation[recommendation['Total Ratings']>100].sort_values('Correlation',ascending=False).reset_index()
```
